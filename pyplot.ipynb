```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 12, 5, 8, 7]
y2 = [5, 4, 6, 3, 9]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines with Legend')
plt.legend()
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
class SubPlot:
    def GraphPlots():
        x_axis=np.random.randint(10,50,8)
        y_axis=np.random.randint(10,50,8)
        plt.subplot(2,2,1)
        plt.grid(color="green")
        plt.plot(x_axis,marker="o",mfc="red",color="blue")
        plt.plot(y_axis,marker="o",mfc="purple",color="black")
        

        a_axis=np.random.randint(10,40,8)
        b_axis=np.random.randint(10,30,8)
        plt.subplot(2,2,2)
        plt.grid()
        plt.barh(a_axis,b_axis)
        
SubPlot.GraphPlots()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 50, 10, 15]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

plt.show()
```
