```python
import numpy as np
import random
import matplotlib.pyplot as plt

class subPlot:
    def GraphPlots():
        x_axis = np.random.randint(10,50,10)
        y_axis = np.random.randint(10,50,10)
        plt.subplot(2,2,1)
        plt.grid(color="green")
        plt.plot(x_axis,y_axis,marker="o",mfc="red",color="blue")

        a_axis = np.random.randint(10,40,8)
        b_axis = np.random.randint(10,30,8)
        plt.subplot(2,2,2)
        plt.barh(a_axis,b_axis)

        c_axis = np.random.randint(50,100,12)
        d_axis = np.random.randint(50,100,12)
        plt.subplot(2,2,3)
        plt.scatter(c_axis,d_axis,color="red")
        plt.grid()

        e_axis = np.random.randint(10,50,5)
        labelValue = ['A','B','C','D','E']
        plt.subplot(2,2,4)
        #f_axis = np.random.randint(10,50,10)
        plt.pie(e_axis,labels=labelValue,shadow=True)
        plt.legend()

        plt.show()

subPlot.GraphPlots()
        
```
